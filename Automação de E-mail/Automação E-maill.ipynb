# Automatizando o envio de e-mail¶
- abrir uma aba ( cntrl + T )
- escrever enderenço e 'enter'
- clicar no botão
- preencher o destinatario, tab
- preencher o corpo email
- clicar ENVIAR
- Não Precisa fazer NADA o script irá fazer TUDO !

* PARA RODAR O SCRIPT: È so clicar 2 vezes nas ESCRITAS e apertar CNTRL + ENTER ou RUN


```python
import pyautogui
import pyperclip
```

```python
# CONFIGURANDO a Pausa
# Voce podera configurar sua pausa é só trocar o número de sua preferencia
pyautogui.PAUSE = 3

# Abrindo uma NOVA ABA
pyautogui.hotkey('ctrl', 't')

# Escrevendo enderenço de E-mail e enter
pyperclip.copy('www.gmail.com')
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('enter')

# Clicar 
pyautogui.click(x=96, y=188)

# Preenchendo o Destinatario 
# Voce podera mudar o email, só não apaguem os ASPAS ''
pyperclip.copy('email@gmail.com')
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('tab')

#Preencher assunto
# Voce podera mudar o assunto do email, só não apaguem os ASPAS ''
pyperclip.copy('Ànalises Dados')
pyautogui.hotkey('ctrl', 'v')
pyautogui('tab')


#Preencher o Corpo do Email
# Voce podera mudar o corpo do email, sá não apaguem os ASPAS ''
mensagem = f"""
Prezado Gestor, 

Aqui nesse E-mail, seguem as Ànalises dos Dados dos últimos seis meses, como solicitados:

Cotação/Ação:{codigo}

Cotação Máxima: {round(maxima, 2)}
Cotação Mínima: {round(minima, 2)}
Cotação Atual: {round(atual, 2)}

Sigo à Disposição!
ATT.
"""

pyperclip.copy(mensagem)
pyautogui.hotkey('ctrl','v')

#Enviar a mensagem
pyautogui.hotkey('ctrl', 'enter')

```
